```python
import socket
import string
import keyboard
import time
import pyautogui
import random
import os

# Conexión a Twitch
def twitch_connect():
    user = os.getenv('TWITCH_USERNAME')
    key = os.getenv('TWITCH_OAUTH_TOKEN')
    s = socket.socket()
    s.connect(('irc.twitch.tv', 6667))
    s.send(f'PASS {key}\r\n')
    s.send(f'NICK {user}\r\n')
    s.send(f'JOIN #{user}\r\n')
    return s

# Leer mensajes del chat de Twitch
def twitch_recieve_messages(s):
    response = s.recv(1024).decode('utf-8')
    if response == 'PING :tmi.twitch.tv\r\n':
        s.send('PONG :tmi.twitch.tv\r\n'.encode('utf-8'))
    else:
        username, message = get_message(response)
        return username, message.strip()

# Extraer el nombre de usuario y el mensaje
def get_message(response):
    parts = response.split(":")
    username_detail = parts[1].split("!")
    username = username_detail[0]
    message = parts[2]
    return username, message

# Funciones para los movimientos y habilidades en Overwatch
def move_forward():
    keyboard.press_and_release('w')

def move_backward():
    keyboard.press_and_release('s')

def move_left():
    keyboard.press_and_release('a')

def move_right():
    keyboard.press_and_release('d')

def ability_one():
    keyboard.press_and_release('shift')

def ability_two():
    keyboard.press_and_release('e')

def ultimate_ability():
    keyboard.press_and_release('q')

def look_left():
    pyautogui.move(-100, 0)

def look_right():
    pyautogui.move(100, 0)

def look_up():
    pyautogui.move(0, -100)

def look_down():
    pyautogui.move(0, 100)

# Conexión a Twitch
s = twitch_connect()

while True:
    username, message = twitch_recieve_messages(s)

    # Añadimos una probabilidad de 1 en 5 para cada comando
    if random.randint(1, 5) == 1:
        if message == '!adelante':
            move_forward()
        elif message == '!atras':
            move_backward()
        elif message == '!izquierda':
            move_left()
        elif message == '!derecha':
            move_right()
        elif message == '!habilidad1':
            ability_one()
        elif message == '!habilidad2':
            ability_two()
        elif message == '!ultimate':
            ultimate_ability()
        elif message == '!mirarizquierda':
            look_left()
        elif message == '!mirarderecha':
            look_right()
        elif message == '!mirararriba':
            look_up()
        elif message == '!mirarabajo':
            look_down()
```
